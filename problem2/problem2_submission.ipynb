```python
# import required modules
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.svm import NuSVR, SVR
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.decomposition import PCA
```

```python
# open files
with open('X_train.npy', 'rb') as f:
    X_train = np.load(f)
with open('y_train.npy', 'rb') as f:
    y_train = np.load(f)
with open('X_test.npy', 'rb') as f:
    X_test = np.load(f)
```

```python
X_train_split, X_val, y_train_split, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=42)

NuSVR_final = NuSVR(C=289, gamma=1e-7, kernel='rbf', nu=0.725)

NuSVR_final.fit(X_train,y_train)
y_test = NuSVR_final.predict(X_test)

sample_submission = pd.read_csv('sample_submission.csv')

for i in range(10000) :
    sample_submission['label'][i] = y_test[i]

csv_file_name_nu = 'nusvr_tuned_final.csv'
sample_submission.to_csv(csv_file_name_nu, index=False)
```

```python
X_train_split, X_val, y_train_split, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=42)

SVR_final = SVR(C=121, epsilon=0.0025, gamma=1.2e-07, kernel='rbf')

SVR_final.fit(X_train,y_train)
y_test = SVR_final.predict(X_test)

sample_submission = pd.read_csv('sample_submission.csv')

for i in range(10000) :
    sample_submission['label'][i] = y_test[i]

csv_file_name = 'svr_tuned_final.csv'
sample_submission.to_csv(csv_file_name, index=False)
```
